## 条件判断
- 缩进一般为**单个制表符/两个空格/四个空格（注意tab和手摁四个空格的区别）**
- 同一层代码的**缩进是相同的**

if 条件1为真

elif 条件2为真  
[space] * 4    执行此代码  
……
else 以上条件均为假

```python
### 条件判断示例
### 程序随机产生一个0~300之间的整数，玩家竞猜，若猜中则提示Bingo，若猜大了提示Toolarge，否则提示Too small
from random import randint
x = randint(0, 300)
digit = int(input('Please input a number between 0~300: '))
if digit == x :
    print('Bingo!')
elif digit > x:
    print('Too large, please try again.')
else:
    print('Too small, please try again.')
```

```python
### 将输入的以分钟为单位的时间显示转换为以xx天xx小时xx分钟显示

a = int(input("请输入分钟数:"))

if a < 60:
    print('a' + 'mins')
elif a//60 < 24 and a >= 60:
    b = a%60
    c = a//60
    print('{}hours {}mins'.format(c, b))
elif (a//60 >= 24) & ((a//60)//24 < 365):
    b = a%60
    c = (a//60)%24
    d = (a//60)//24
    print('{}days {}hours {}mins'.format(d, c, b))
elif ((a//60)//24) >=365:
    b = a%60
    c = (a//60)%24
    d = ((a//60)//24)%365
    e = ((a//60)//24)//365
    print('{}years {}days {}hours {}mins'.format(e, d, c, b))
```

## 循环While语句

- 循环结构是满足一个指定的条件，每次使用不同的数据对算法中的计算或处理步骤完全相同的部分重复计算若干次的算法结构，也称为重复结构
- 语法结构：  
    while 表达式为真：
        执行循环体
    直到表达式的值为False或发生异常时停止循环
    

```python
### 计算1-100之间所有数字之和
i = 1
s = 0
while i <= 100:
    s += i
    i += 1
print(s)
```

```python
### 求两个正整数的最大公约数和最小公倍数
### S1：判断x除以y的余数r是否为0。若r为0则y是x、y的最大公约数，继续执行后续操作；否则y→x，r→y重复执行第S1步。S2：输出（或返回）y。（辗转相除法）

x = eval(input('first number:'))
y = eval(input('second number:'))
z = x * y
if x < y:
    x,y=y,x
    while x%y !=0:
        r = x%y
        x = y 
        y = r
    print('最大公约数=', y)
    print('最小公倍数=', z//y)
```

## 异常（Exception）处理语句

- python中的每一类异常都是类的实例/用异常对象（exception object）表示异常情况

```python
dir(__builtins__)    ### 查看异常类
```

```python
### 异常处理语法
try:                 #被检测语句，当程序不报错时运行该代码块
    num1=int(input('Please enter a number:'))
    num2=int(input('Please enter another number:'))
    print(num1//num2)
except:               #异常处理，如果出现某种错误则执行该代码块；错误类型可选：except Error type
    print('please enter a digit!')
```

```python
### 异常处理语法
try:
    num1=int(input('Please enter a number:'))
    num2=int(input('Please enter another number:'))
    print(num1//num2)
except ValueError as err:
    print('please enter a digit!')
    print(err)
except ZeroDivisionError:     ### 多个except字句可捕捉多个异常
    print('zero cannot be divided!')
```

```python
### 空except语句和as
try:
    num1=int(input('Please enter a number:'))
    num2=int(input('Please enter another number:'))
    print(num1//num2)
except Exception as err:    ### 当无法预知报错类型时可输出错误原因
    print(err)
```

```python
### 异常处理的else语句
try:
    num1=int(input('Please enter a number:'))
    num2=int(input('Please enter another number:'))
    print(num1/num2)
except Exception as err:
    print(err)
else:     ### 当try模块的语句执行不报错时，执行该代码块
    print('everything is fine')
    
```

```python
### 反复运行知道不报错 ---> 循环结构
while True:
    try:
        num1=int(input('Please enter a number:'))
        num2=int(input('Please enter another number:'))
        print(num1/num2)
        break
    except Exception as err:
        print(err)
```

## Break, Continue, finally 语句
- break语句：break语句终止当前循环，转而执行循环之后的语句
- 在while和for循环中，continue语句的作用：
   - 跳过循环体内continue后面的语句，并开始新的一轮循环
   - while循环则判断循环条件是否满足
   - for循环则判断迭代是否已经结束
- finally语句：在异常处理中，无论是否报错都执行该部分代码块

```python
### Example
### break 语句
### 输出2~100中的素数，每行显示5个
from math import sqrt
j = 2
count = 0
while j <= 100:
    i = 2
    k = int(sqrt(j))
    while i<=k:
        if j%i == 0:
            break
        i += 1
    if i > k:
        count += 1
        if count%5 == 0:
            print(j, end='\n')
        else:
            print(j, end=' ')
    j += 1
```

```python
### break 语句和异常处理模块1
while True:
    try:
        num1=int(input('Please enter a number:'))
        num2=int(input('Please enter another number:'))
        print(num1/num2)
    except Exception as err:
        print(err)
    else:
        break  ### 如果try部分未报错，则跳出循环

```

```python
### break 语句和异常处理模块2
alist = [1,2,3,4,5]
i = 0
while True:
    try:
        print(alist[i])
    except IndexError:
        print('index error')
        break    ### 如果报错则跳出循环
    else:
        i += 1
```

```python
### continue 语句
for i in range(1,21):
    if i % 3 != 0:
        continue    ### continue跳出本轮循环
    print(i, end = ' ')
```

```python
for i in range(1,21):
    if i % 3 !=0:
        break
    print(i ,end=' ')   ### break 跳出所有循环
```

```python
### 异常处理中的finally语句：无论try部分的内容是否报错均执行该代码块
def finallytest():
    try:
        num1=int(input('Please enter a number:'))
        num2=int(input('Please enter another number:'))
        print(num1/num2)
    except Exception as err:
        print(err)
    finally:
        print('haha')
```

## 列表解析
列表解析中的多个for语句相当于是for结构的嵌套使用  
[ 表达式 for 表达式1 in 序列1 for 表达式2 in 序列2 … for 表达式N in 序列N if 条件 ]

```python
[x for x in range(0,9)]
```

```python
[x**2 for x in range(10)]
```

```python
[x ** 2 for x in range(10) if x ** 2 < 50]
```

```python
[(x + 1,y + 1) for x in range(2) for y in range(2)]
```

## 常见报错

- BaseException: 所有异常的基类
- Exception：常规异常的基类
- AttributeError：对象不存在此属性
- IndexError：序列中无此索引
- IOError：输入/输出操作失败
- KeyboardInterrupt：用户中断执行（通常输入ctrl-c）
- KeyError：映射中不存在此键
- NameError：找不到名字（变量）
- SyntaxError：Python语法错误
- TypeError：对类型无效的操作
- ValueError：传入无效的参数
- ZeroDivisionError：除运算的第二个参数为0
